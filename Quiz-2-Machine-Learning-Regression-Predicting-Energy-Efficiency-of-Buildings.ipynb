```python
import pandas as pd
```

```python
df=pd.read_csv('https://archive.ics.uci.edu/ml/machine-learning-databases/00374/energydata_complete.csv')
```

```python
df.head()
```

```python
df1 = df.drop(columns=[ 'date' , 'lights' ])
```

```python
df1.head()
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
```

```python
normalised_df = pd.DataFrame(scaler.fit_transform(df1), columns=df1.columns)
```

```python
normalised_df
```

```python
features_df = normalised_df.drop(columns=['Appliances'])
```

```python
features_df
```

```python
target_df = normalised_df['Appliances']
```

```python
target_df
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(features_df, target_df,test_size= 0.3 , random_state= 42 )
```

```python
from sklearn import linear_model
```

```python
linear_model = linear_model.LinearRegression()
```

```python
linear_model.fit(x_train, y_train)
```

```python
predicted_values = linear_model.predict(x_test)
```

```python
from sklearn.metrics import r2_score
```

```python
r2_score = r2_score(y_test, predicted_values)
```

```python
round(r2_score, 2)
```

```python
from sklearn.metrics import mean_absolute_error
```

```python
mae = mean_absolute_error(y_test, predicted_values)
```

```python
round(mae, 3)
```

```python
import numpy as np
```

```python
rss = np.sum(np.square(y_test - predicted_values))
```

```python
round(rss, 2)
```

```python
from sklearn.metrics import  mean_squared_error
```

```python
rmse = np.sqrt(mean_squared_error(y_test, predicted_values))
```

```python
round(rmse, 3)
```

```python
def weightsfunc(model, feature, column_name):
  #this function returns the weight of every feature
  weights = pd.Series(model.coef_, feature.columns).sort_values()
  weights_df = pd.DataFrame(weights).reset_index()
  weights_df.columns = ['Features', column_name]
  weights_df[column_name].round(3)
  return weights_df
```

```python
linear_model_weights = weightsfunc(linear_model, x_train, 'Linear_Model_Weight')
```

```python
linear_model_weights
```

```python
linear_model_weights.min()
```

```python
linear_model_weights.max()
```
